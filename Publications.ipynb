```python
import requests
from bs4 import BeautifulSoup

# Function to scrape data from a Google search page
def scrape_google_search(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Extract the desired information from the search results page
    # Replace this code with your specific scraping logic
    h2_element = soup.find_all("div", class_="egMi0 kCrYT")
    l = []
    for ele in h2_element:
      h = ele.find('a')
      l.append(h['href'])
    
    filter = []
    for text in l:
      # Extract URLs using regular expressions
      urls = re.findall(r'/url\?q=(.*?)&', text)

      # Filter the desired sites
      filtered_urls = [url for url in urls if any(site in url for site in sites)]

      # Print the filtered URLs
      for url in filtered_urls:
          filter.append(url)
    return filter

# Define the base URL for the Google search
base_url = 'https://www.google.com/search?q=nowadays+bar+in+newyork'

publications = []
# Scrape the first two pages of search results
for page in range(2):
    # Construct the URL for the current page
    if page == 0:
        url = base_url
    else:
        url = f'{base_url}&start={page * 10}'
    
    # Scrape data from the current page
    titles = scrape_google_search(url)
    publications += titles
```
